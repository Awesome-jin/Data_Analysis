```python
#2개 컬럼의 관계를 파악하기 위한 산포도

df = pd.read_csv('./data/auto-mpg.csv')
#열 이름 지정

df.columns = ['mpg', 'cylinder', 'displacement','horsepower','weight','acceleration','yr','origin','name']
plt.scatter(df['weight'],df['mpg'], s=df['cylinder']*5) #3차원이 아니더라도 변수 3개의 관계를 2차원으로 확인 가능하게 하는 원크기 옵션 s
plt.title('중량과 연비 관계')
plt.xlabel('연비', size=10)
plt.ylabel('mpg', size=10)
plt.show()
```

# 파이 그래프

```python
df = pd.read_csv('./data/korea.csv', encoding='cp949')
explode = []

for i in range(0,20,1):
    explode.append(0)
explode[2] = 0.3

plt.figure(figsize=(15,5))
plt.pie(df['점수'],labels=df['이름'],explode=explode, autopct='%1.1f%%')
plt.title('학생별 기여도')
plt.legend()
plt.show()
```
