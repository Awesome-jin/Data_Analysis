## seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sb

#seaborn에는 기본 데이터셋이 몇개 있다. tips라는 데이터를 호출해보자
tips = sb.load_dataset('tips')
#type(tips)
#tips.head()
#tips.info()

fig = plt.figure(figsize=(8,6))

#1행 2열 1번째 영역
#ax1 = fig.add_subplot(1,2,1)
#ax2 = fig.add_subplot(1,2,2)
#sb.set_style('darkgrid')

#데이터에 데이터프레임을, x,y에 컬럼이름 설정 
#sb.regplot(x='total_bill', y='tip', data=tips, fit_reg=True, ax=ax1)
#sb.regplot(x='total_bill', y='tip', data=tips, fit_reg=False, ax=ax2)

#3개 컬럼의 관계 보기 = lmplot
#x, y, hue 옵션 이용
#hue에는 categorical로 설정 가능
sb.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', fit_reg=True)
```

### tip 컬럼 분포 확인

```python
fig = plt.figure(figsize=(8,6))
sb.set_style('darkgrid')
sb.distplot(tips['tip'])
plt.show()
#>> 히스토그램은 측정하지 않은 level의 표현이 불가능하다

```

## heatmap

```python
flight = sb.load_dataset('flights')
#flight.info()
#flight.head()

#1. heatmap을 만들기 전 제일 먼저 해야 하는 작업 : pivot 테이블 만들기
pivot = flight.pivot('month','year','passengers') #1-1. 축 2개를 앞에. 보고자하는 데이터를 마지막에 기입
#print(pivot) : 제대로 피벗테이블이 그려졌는지 확인
plt.figure(figsize=(10,8))
sb.heatmap(pivot, annot=True, fmt='d')   #annot : 데이터 화면에 출력할지 여부. fmt : 숫자 출력 포맷 d는 decimal
plt.show()
```
