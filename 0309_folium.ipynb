```python
import folium
import pandas as pd
from matplotlib import font_manager, rc
import platform
#매킨토시의 경우
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
#윈도우의 경우
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
#folium.Marker(location=[37.572656, 126.973304], popup='KB', icon=folium.Icon(icon='cloud')).add_to(m)
m #지도 출력 바로 가능
#m.save('map.html') : #지도를 저장하는 과정인데 Jupyter에선 굳이 필요 없음

df = pd.read_excel('./data/서울지역 대학교 위치.xlsx')
#print(df.head()) 제대로 읽어졌는지 확인
m = folium.Map(location=[37.55, 126.97], zoom_start=12)
#m

#데이터프레임의 각 셀의 값을 행 단위로 읽기
#zip은 여러개의 데이터를 하나로 묶어주는 역할 - 튜플로 묶어줌
#하나로 묶은 데이터를 다시 name, lat, lng로 나누어서 읽는 것
for name, lat, lng in zip(df['Unnamed: 0'],df['위도'], df['경도']): #Unnamed: 0   << 컬럼이름이 없을 때 쓰는 명령어 : 뒤 공백 주의!!
    #folium.Marker(location=[lat,lng],popup=name,icon=folium.Icon(icon='cloud')).add_to(m)
    folium.CircleMarker(location=[lat,lng],popup=name,radius=10, color='green', fill=True, fill_color='coral',fill_opacity=0.7).add_to(m)
    #학생 수 데이터가 있다면 radius의 값을 학생수로 넣어주면 학생수별로 원의 크기가 달라지는 효과도 표시 가능 할 듯!
    m #한글이 깨질 수 있음 (저장해서 html로 보면 제대로 보임)
#m.save('map.html')
```
